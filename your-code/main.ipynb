#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'

response = requests.get(url)
response
````

````python
# your code here
soup = BeautifulSoup(response.content)
langs = soup.find_all('div', {'class': 'central-featured-lang'})
langs
````

````python
languages = [line.strong.text for line in langs]
print(languages)
articles = [line.small.text.replace('\xa0', '.') for line in langs]
print(articles)
````

````python
langs_dict = {}

for k, v in zip(languages,articles):
    langs_dict[k] = v
    
langs_dict
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code here

response = requests.get(url)
response
````

````python
soup = BeautifulSoup(response.text, 'html.parser')

# Find all movie containers
movie_containers = soup.find_all('div', class_='lister-item mode-detail')

# Create a list to store the scraped data
data = []

# Loop through all movie containers
for movie in movie_containers:
    # Find the movie name and release year
    name = movie.find('h3', class_='lister-item-header').a.text
    year = movie.find('span', class_='lister-item-year text-muted unbold').text
    # Find the movie summary
    summary = movie.find('p', class_='text-muted').text

    # Store the data in a list
    data.append([name, year, summary])

# Create a pandas dataframe from the scraped data
movies = pd.DataFrame(data, columns=['Movie Name', 'Year', 'Summary'])

# Randomly display 10 rows of the dataframe
movies.sample(n = 10)


#I can't find the summary in this specific page, I'm thinking we would have to iterate through the links 
#to find it. I'll confirm it with Inês and submit it again.
````
